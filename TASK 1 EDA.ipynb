```python
#Import Library
import seaborn as sns
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import pandas as pd # data processing
```

```python


# 1. Distribution of Sales Amount
plt.figure(figsize=(8, 6))
sns.histplot(df['Sales Amount'], kde=True, bins=10, color='green')
plt.title('Distribution of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()

```

```python
# 2. Distribution of Profit
plt.figure(figsize=(8, 6))
sns.histplot(df['Profit'], kde=True, bins=10, color='blue')
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python

# 3. Count of Products Sold
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='Product', order=df['Product'].value_counts().index, palette='viridis')
plt.title('Count of Products Sold')
plt.xlabel('Count')
plt.ylabel('Product')
plt.show()
```

```python
# 4. Sales Count by Region
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Region', palette='magma')
plt.title('Sales Count by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python


# 1. Sales Amount vs Profit (Scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Sales Amount', y='Profit', hue='Region', palette='viridis')
plt.title('Sales Amount vs Profit by Region')
plt.xlabel('Sales Amount')
plt.ylabel('Profit')
plt.legend(title='Region')
plt.show()

```

```python

# 2. Average Sales Amount by Product (Bar plot)
avg_sales_product = df.groupby('Product')['Sales Amount'].mean().sort_values()
plt.figure(figsize=(8, 6))
avg_sales_product.plot(kind='bar', color='teal')
plt.title('Average Sales Amount by Product')
plt.xlabel('Product')
plt.ylabel('Average Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Average Profit by Region (Bar plot)
avg_profit_region = df.groupby('Region')['Profit'].mean().sort_values()
plt.figure(figsize=(8, 6))
avg_profit_region.plot(kind='bar', color='orange')
plt.title('Average Profit by Region')
plt.xlabel('Region')
plt.ylabel('Average Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Sales Amount by Region (Boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Region', y='Sales Amount', palette='magma')
plt.title('Sales Amount by Region')
plt.xlabel('Region')
plt.ylabel('Sales Amount')
plt.show()
```
