```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values[missing_values > 0])
```

```python
# Fill NaN values in numeric columns with their respective column mean
df[df.select_dtypes(include=["number"]).columns] = df.select_dtypes(include=["number"]).apply(lambda col: col.fillna(col.mean()))


```

```python
# Fill NaN values in numeric columns with their respective column medians
df[df.select_dtypes(include=["number"]).columns] = df.select_dtypes(include=["number"]).apply(lambda col: col.fillna(col.median()))

```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Analyze Numerical Columns
# Select numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns

# Univariate analysis for numerical columns
for col in numerical_cols:
    print(f"Analysis of {col}:")
    print(df[col].describe())  # Summary statistics
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=20, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Analyze Categorical Columns
# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Univariate analysis for categorical columns
for col in categorical_cols:
    print(f"Value Counts for {col}:")
    print(df[col].value_counts())  # Value counts for each category
    plt.figure(figsize=(8, 5))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
#Boxplots for Numerical Variables
# Boxplot for numerical variables to detect outliers
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(y=df[col], color='cyan')
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

```python
#Analyze Relationships Between Numerical Variables
# Select numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns

# Pairwise scatterplots for numerical variables
sns.pairplot(df[numerical_cols], diag_kind='kde')
plt.suptitle('Pairwise Scatterplots', y=1.02)
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Relationship Between Numerical and Categorical Variables
# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Boxplots for numerical variables against categorical variables
for cat_col in categorical_cols:
    for num_col in numerical_cols:
        plt.figure(figsize=(8, 5))
        sns.boxplot(x=df[cat_col], y=df[num_col], palette='Set2')
        plt.title(f'Boxplot of {num_col} by {cat_col}')
        plt.xlabel(cat_col)
        plt.ylabel(num_col)
        plt.show()
```
